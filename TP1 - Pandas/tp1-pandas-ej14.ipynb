# Inicialización de los dataframes

```python
# Nota: cargo la columna apps para poder dropear duplicados después
ps_apps = pd.read_csv("googleplaystore.csv", skiprows=[10473], usecols=["App", "Category", "Reviews"])

ps_apps = ps_apps.drop_duplicates(subset=["App"], keep="first")

ps_apps["Category"] = ps_apps["Category"].astype("category")
```
